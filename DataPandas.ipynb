# Mengubah string menjadi date

```python
date_set = sns.load_dataset("taxis")
date_set["pickup"] = pd.to_datetime(date_set["pickup"])
date_set["dropoff"] = pd.to_datetime(date_set["dropoff"])
data_set.info()
```

```python
date_set["dropoff"].dt.month.unique()
```

```python
# mendefinisikan time zone mengconvert
date_set["dropoff"].dt.tz_localize("UTC").dt.tz_convert("Asia/Jakarta")
```

```python
# menggubah menjadi priode
date_set["pickup"].dt.to_period("1m")
```

```python
(date_set["pickup"] - date_set["dropoff"])
```

```python
(date_set["dropoff"] - date_set["pickup"]).dt.seconds //60
```

```python
date_set["dropoff"] + pd.to_timedelta(4 , unit= "H")
```

```python
# rata yg ada di data dropoff
date_set["dropoff"].mean()
```

```python
date_set.groupby("payment").resample("1d", on = "pickup").agg({"passengers" : "max",
                                                              "tip" : "max"})
```

```python
# membuat data_set baru dengan fungsi leg (-) dan leag(+)
date_set["pickup_hmin1"] = date_set["pickup"].shift(1)
```

```python
date_set.filter(like = "pickup")
```

```python
# membandingkan anatara series pickup_hmin1 dan pickup
date_set["pickup"] - date_set["pickup_hmin1"]
```

```python
# menggrutukan data
date_set.sort_values("pickup")
```

```python
# menggunakan rolling
date_set.sort_values("pickup")["passengers"].rolling(3).max()
```
